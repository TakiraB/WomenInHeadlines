###imports

```python
import pandas as pd
```

```python
# file dowloaded from: https://www.kaggle.com/datasets/thedevastator/women-in-headlines-bias?resource=download&select=headlines.csv
headlineframe = pd.read_csv('headlines.csv')
```

##Cleaning the data

```python
headlineframe
```

```python
# filter out the columns we want
headlines = headlineframe[['site', 'headline_no_site','time','country', 'bias']]

# drop NaNs
headlines = headlines.dropna()
```

```python
headlines
```

```python
# checking values
headlines['site'].unique()
```

```python
# working with a copy
copy = headlines.copy()
```

```python
copy
```

```python
# creating new columns 
copy['date'] = " "
copy['timezone'] = " "

# spliting the columns 
copy[['date', 'time']] = copy.time.str.split(" ", expand = True)
copy[['time', 'timezone']] = copy.time.str.split("+", expand = True)

# dropping unwanted info
copy = copy.drop(columns='timezone')
```

```python
# rounding just because
copy['bias'] = round(copy['bias'], 2)
```

```python
# renaming
copy = copy.rename(columns={'headline_no_site': 'headline'})
```

```python
copy
```

```python
# converting to datetime
copy['time'] = pd.to_datetime(copy['time'])
copy['date'] = pd.to_datetime(copy['date'])
```

```python
# reformatting
copy['date'] = copy['date'].dt.strftime("%m/%d/%Y")
copy['time'] = copy['time'].dt.strftime("%H:%M")
```

```python
copy
```

```python
# reorganizing
copy = copy[['date', 'time', 'headline','site','country', 'bias']]
```

```python
# making new columns for split
copy['s2'] = " "
copy['s3'] = " "

# getting the site title
copy[['site', 's2', 's3']] = copy.site.str.split(".", expand = True)

# dropping unwanted info
copy = copy.drop(columns=['s2', 's3'])
```

```python
copy.info()
```

```python
# converting back to datetime
copy['date'] = pd.to_datetime(copy['date'])
```

```python
copy.info()
```

##Save to cvs

```python
copy.to_csv('filtered_headlines.csv', index_label=False)
```
